```python
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px
```

```python
df = pd.read_csv('data_to_map.csv')
```

```python
df.head()
```

```python
mapbox_token = 'pk.eyJ1Ijoic3RldmllY2wiLCJhIjoiY2w3OWgwaGgzMGRkdzNwbzQxb2IwY2N6dyJ9.Ws3kDqCyqYZ4k3LI-ah1pA'
coord_med = [6.217, -75.567]
fig = go.Figure(go.Scattermapbox(
    lat=df['latitud'],
    lon=df['longitud'],
    mode='markers',
    text=df[['sexo', 'medio_transporte', 'modalidad']],
    marker=go.scattermapbox.Marker(size=9,
                                   opacity=0.5
                                   )
))

fig.update_layout(
    mapbox_style='open-street-map',
    autosize=True,
    hovermode='closest',
    mapbox=dict(
        accesstoken=mapbox_token,
        bearing=0,
        center=dict(
            lat=coord_med[0],
            lon=coord_med[1],
        ),
        zoom=10
    )
    
    )

fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
```

```python
sexo_ = ['Hombre', 'Mujer']
medio_transporte_ = ['Autobus', 'Taxi', 'Metro']
comuna_ = ['Aranjuez']
df_filtro = df[(df.sexo.isin(sexo_)) &
                (df.medio_transporte.isin(medio_transporte_)) &
                (df.sede_receptora.isin(comuna_))]
df_filtro['color'] = df_filtro.sexo.replace(to_replace={'Hombre':'blue', 'Mujer':'red'})
mapbox_token = 'pk.eyJ1Ijoic3RldmllY2wiLCJhIjoiY2w3OWgwaGgzMGRkdzNwbzQxb2IwY2N6dyJ9.Ws3kDqCyqYZ4k3LI-ah1pA'
coord_med = [6.217, -75.567]
fig = go.Figure(go.Scattermapbox(
    lat=df_filtro['latitud'],
    lon=df_filtro['longitud'],
    mode='markers',
    text=df_filtro[['sexo', 'medio_transporte', 'modalidad']],
    fillcolor='yellowgreen',
    marker=go.scattermapbox.Marker(size=9,
                                   opacity=0.5,
                                   color=df_filtro['color']
                                   )
))

fig.update_layout(
    mapbox_style='open-street-map',
    autosize=True,
    hovermode='closest',
    mapbox=dict(
        accesstoken=mapbox_token,
        bearing=0,
        center=dict(
            lat=coord_med[0],
            lon=coord_med[1],
        ),
        zoom=10
    )
    
    )

fig.update_layout(margin={"r":0,"t":0,"l":0,"b":0})
fig.show()
```
