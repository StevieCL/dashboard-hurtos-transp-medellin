```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
```

```python
df = pd.read_csv('data_dash.csv')
```

```python
def pie_chart():
    medio_trans = df.medio_transporte.value_counts()
    names = medio_trans.index.to_list()
    values = list(medio_trans.values)
    fig = px.pie(values=values, names=names,
        hole=.3, title='Medio de transporte')
    fig.update_traces(textposition='inside', textinfo='percent+label')
    fig.update_layout(
        title={
            'text':'Medio de transporte',
            'y':0.95,
            'x':0.48,
            'xanchor':'center',
            'yanchor':'top'
        }
    )
```

```python
top_mod = df.modalidad.value_counts().head(3)
fig = px.bar(x=top_mod.index, y=top_mod.values,
             labels={
                 'x':'Modalidad de robo',
                 'y':'Total'
             })
fig.update_layout(
    title={
        'text':'Top 3 Modalidad de robo',
        'y':0.95,
        'x':0.5,
        'xanchor':'center',
        'yanchor':'top'
    }
)
```

```python
px.histogram(df, x='edad', color='sexo')
```

```python
fig = go.Figure()
fig.add_trace(go.Histogram(x=df[df.sexo == 'Hombre']['edad'], name='Hombre'))
fig.add_trace(go.Histogram(x=df[df.sexo == 'Mujer']['edad'], name='Mujer'))

# Overlay both histograms
fig.update_layout(barmode='overlay')
# Reduce opacity to see both histograms
fig.update_traces(opacity=0.75)
fig.show()
```
